```python
chinese = int(input("國文幾分"))
math = int(input("數學幾分"))
english = int(input("英文幾分"))

total = chinese + math + english
print("總分",total,"分",sep="")
average = round(total/3,ndigits=2)
print("平均",average,"分",sep="")
```
